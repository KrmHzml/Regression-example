## 6-Feature Engineering

### 6.1-Torque

Torque is the rotational force of the engine; It enables the wheels of the vehicle to be converted into thrust. So in a way, it is related to the traction power of the vehicle.

Therefore, cars with higher Torque can be expected to be more expensive.

Torque = 5252 x HP/ RPM

```python
# insert 'torque' column

data.insert(10, 'torque', 
            pd.Series(data.horsepower * 5252 / data.peakrpm, index=data.index))
```

```python
plt.title('Torque - Price', fontsize=16)

sns.regplot(x=data.torque, y=data.price)

plt.show()
```

```python
plt.title('Torque Distribution', fontsize=18)
sns.distplot(data.torque)

plt.show()
```

### 6.2-Full Economy

Average fuel consumption of the car in and outside the city.

```python
#insert 'fueleconomy' column
data['fueleconomy'] = (0.55 * data.citympg) + (0.45 * data.highwaympg)
```

```python
## Lastly drop the unnecessary columns
data.drop(columns=['car_ID','manufacturer','doornumber','carheight',
                   'compressionratio', 'symboling','stroke','citympg',
                   'highwaympg', 'fuelsystem', 'peakrpm'],
                    axis=1,inplace=True)
```

## 10-Multiple Linear Regression

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

### 10.1-RFE (Recursive Feature Elimination):

It runs backwards Multiple Linear Regression, reducing one variable at a time (backward elimination).

By eliminating in this way, it comes up to the number of **n_features_to_select** given as a parameter.

Eliminates the least important in each elimination. (i.e. the one with the largest p-value)

```python
rfe = RFE(estimator=lr,n_features_to_select=10)
```

```python
rfe = rfe.fit(X_train,y_train)
```

Attributes of RFE:

**ranking_**: Selection rank, with the highest importance from the first (rank=1).

**support_**: Is the variable selected -> True, False.

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#only the selected columns
X_train.columns[rfe.support_]
```

```python
# So now we know the important columns
X_train_rfe = X_train[X_train.columns[rfe.support_]]
```

```python
X_train_rfe 
```

### 10.2-OLS Analysis

```python
#copy for OLS
X_train_rfemodel=X_train_rfe.copy()
```

```python
# add_constant for statsmodels -> column of 1s for beta_0
X_train_rfemodel = sm.add_constant(X_train_rfemodel)
```

```python
#Execute OLS
lr=sm.OLS(y_train,X_train_rfemodel).fit()
```

```python
print(lr.summary())
```

The first thing we see here is that the p-value of the variable named **dochv** is greater than 0.05.

So it's unimportant, we can drop it.

```python
#OLS function
def train_ols(X,y):
    X = sm.add_constant(X)
    lr = sm.OLS(y,X).fit()
    print(lr.summary())
```

```python
X_train_rfemodel = X_train_rfemodel.drop(['dohcv'],axis=1)
train_ols(X_train_rfemodel, y_train)
```

Now the variable **five** is unimportant.

Let's drop it and run OLS again:

```python
X_train_rfemodel = X_train_rfemodel.drop(['five'],axis=1)
train_ols(X_train_rfemodel, y_train)
```

And,also **two**

```python
X_train_rfemodel = X_train_rfemodel.drop(['two'],axis=1)
train_ols(X_train_rfemodel, y_train)
```
